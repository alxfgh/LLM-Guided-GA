```python
import copy

import group_selfies as sf
from rdkit.Chem import AllChem as Chem
import rdkit.Chem.Draw as Draw
import numpy as np
import random 
import copy
from tqdm import tqdm
import time
import numpy as np
from group_selfies import GroupGrammar


global grammar
grammar = GroupGrammar.essential_set()
```

```python
import csv
i = 0
# open the input and output files
with open('output_5.csv', 'r') as input_file, open('outputSelfie_5.csv', 'w', newline='', encoding='utf-8') as output_file:
    # Skip the header line in the input file
    next(input_file)
    # create a CSV writer object
    writer = csv.writer(output_file)

    # loop over each line in the input file
    for line in input_file:
        i = i + 1
        if i % 10000 == 0:
            print(i)
        try:
            # process the line using the process_line() function
            values = grammar.full_encoder(Chem.MolFromSmiles(line.strip()))
            # write the values to the output file
            writer.writerow([values])
        except:
            # if an error occurs, just skip the line
            print(f"Error in line: {line}")
            continue
```
